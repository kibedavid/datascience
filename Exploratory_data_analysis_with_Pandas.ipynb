1. Demonstration of the main Pandas methods
pandas is useful in working with data stored in table formats such as .csv, .tsv, or .xlsx.

  With pandas you can load process, and analyze data using SQL-like commands.  pandas can be used with Matplotlib and Seaborn, Pandas provides a wealth of opportunities for visualizing and analyzing tabular data.

  core data structures in Pandas are Series and DataFrames.

  A Series is a one-dimensional indexed array of a single data type, while a DataFrame is a two-dimensional table where each column contains data of the same type.

```python
import numpy as np
import pandas as pd
pd.set_option("display.precision",2)
```

```python
DATA_URL = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/"
```

 Main methods in analyzing a dataset on the churn rate of telecom operator clients.

```python
df = pd.read_csv(DATA_URL + "telecom_churn.csv")
df.head()
```

```python
## Over view of the data
print(df.shape)
```

From the output, we can see that the table contains 3333 rows and 20 columns.



```python
##  printing out column names
print(df.columns)
```

```python
## getting the general information of dataframe
print(df.info())
```

There is no missing data  because each column contains 3333 observations, the same number of rows we saw before with shape.

```python
## To  change the column type with the astype method. Let’s apply this method to the Churn feature to convert it into int64
df["Churn"] = df["Churn"].astype("int64")
```

```python
## The describe method shows basic statistical characteristics of each numerical feature (int64 and float64 types): number of non-missing values, mean, standard deviation, range, median, 0.25 and 0.75 quartiles.
df.describe()
```

```python
## In order to see statistics on non-numerical features, one has to explicitly indicate data types of interest in the include parameter.
df.describe(include=["object", "bool"])
```

```python
## For categorical (type object) and boolean (type bool) features we can use the value_counts method. Let’s take a look at the distribution of Churn:
df["Churn"].value_counts()
```

```python
## above we see 2850 users out of 3333 are loyal; their Churn value is 0.
## To calculate fractions, pass normalize=True to the value_counts function.
df["Churn"].value_counts(normalize=True)
```

```python
## SORTING
## Sort Total days charge into descending order

df.sort_values(by="Total day charge", ascending=False).head()
```

```python
## i can also sort using multiple columns
df.sort_values(by =[ "Churn", "Total day charge"], ascending=[True, False]).head()
```

```python
## indexing and retrieving data

## Boolean indexing with one column is also very convenient. The syntax is df[P(df['Name'])], where P is some logical condition that is checked for each element of the Name column. The result of such indexing is the DataFrame consisting only of the rows that satisfy the P condition on the Name column.

## Lets us this to answer afew  questions with the data we have:
```

```python
## What are the average values of numerical features for churned users?

df.select_dtypes(include=np.number)[df["Churn"] == 1].mean()
```

```python
## How much time (on average) do churned users spend on the phone during daytime?
df[df["Churn"] == 1]["Total day minutes"].mean()
```

```python
## What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?
df[(df["Churn"] == 0) & (df["International plan"] == "No")]["Total intl minutes"].max()
```

```python
## second method i will use for indexing is loc index name & iloc index number
df.loc[0:5, "State":"Area code"]
```

```python
df.iloc[0:5, 0:3]
```

Applying Functions to Cells, Columns and Rows

```python
##The apply method can also be used to apply a function to each row.
df.apply(np.max)
```

```python
df[df["State"].apply(lambda state: state[0] == "W")].head()
```

```python
## The map method can be used to replace values in a column by passing a dictionary of the form {old_value: new_value} as its argument
d = {"No": False, "Yes": True}
df["International plan"] = df["International plan"].map(d)
df.head()
```
